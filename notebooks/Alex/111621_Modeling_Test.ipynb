## Redoing the above with types of sentiment (pos/neg/neutral)

**Emotion Categorical Encoding Key:**

0. Negative, "Negative emotion" 
1. None, "No emotion toward brand or product"
2. Positive, "Positive emotion"
3. Unsure, "I can't tell"

```python
def remove_unsure(score):
    if score == 3:
        return 1
    else: 
        return score
```

```python
y_train_multi_class['emotion'] = y_train_multi_class['emotion'].apply(lambda x: remove_unsure(x))
```

```python
y_train_multi_class.value_counts(normalize=True)
```

```python
def has_sentiment(tweet: str) -> bool:
    a_dictionary = {"neg": sia.polarity_scores(tweet)['neg'], 
                    "neu": sia.polarity_scores(tweet)['neu'],
                    "pos": sia.polarity_scores(tweet)['pos']}

    max_key = max(a_dictionary, key=a_dictionary.get)
    
    sentiment_dict = {'neg':0, 'neu':1, 'pos':2}

    return sentiment_dict[max_key]
```

```python
y_preds_multi = X_train['cleaned_tweets'].apply(lambda x: has_sentiment(str(x)))
```

### Creating some new sentiment based features to dump in classifiers, all of this is from the article linked at the top of the notebook

```python
def extract_features(text: str) -> bool:
    features = dict()
    wordcount = 0
    compound_scores = list()
    positive_scores = list()
    
#     print(text)
    
    compound_scores.append(sia.polarity_scores(text)['compound'])
    positive_scores.append(sia.polarity_scores(text)['pos'])
            
    # Adding 1 to the final compound score to always have positive numbers
    # since some classifiers you'll use later don't work with negative numbers.
    features["mean_compound"] = np.mean(compound_scores) + 1
    features["mean_positive"] = np.mean(positive_scores)

    return features
        
```

```python
def mean_compound(text: str) -> bool:

    return (sia.polarity_scores(text)['compound'])
```

```python
def mean_positive(text: str) -> bool:
    
    return (sia.polarity_scores(text)['pos'])

        
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vec = TfidfVectorizer()
```

```python
X = tfidf_vec.fit_transform(X_train['cleaned_tweets'])
```

```python
tfidf_train_df = pd.DataFrame(X.toarray(), 
                              columns=tfidf_vec.get_feature_names(), 
                              index=X_train.index)
```

```python
tfidf_train_df
```

```python
tfidf_train_df['compound'] = X_train['mean_compound']
tfidf_train_df['positive'] = X_train['mean_positive']
```

```python
from sklearn.naive_bayes import (
    BernoulliNB,
    ComplementNB,
    MultinomialNB,
)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
classifiers = {
    "BernoulliNB": BernoulliNB(),
    "ComplementNB": ComplementNB(),
    "MultinomialNB": MultinomialNB(),
    "KNeighborsClassifier": KNeighborsClassifier(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "RandomForestClassifier": RandomForestClassifier(),
    "LogisticRegression": LogisticRegression(),
    "MLPClassifier": MLPClassifier(max_iter=1000),
    "AdaBoostClassifier": AdaBoostClassifier(),
}
```

```python
for name, sklearn_classifier in classifiers.items():
     classifier = nltk.classify.SklearnClassifier(sklearn_classifier)
     classifier.train(tfidf_train_df)
     accuracy = nltk.classify.accuracy(classifier, tfidf_train_df)
     print(F"{accuracy:.2%} - {name}")
```
